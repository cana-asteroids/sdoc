# Export database to H5F file and replacing the database with new H5F 



```python
import sdoc
```

### Export the database SDOC into an H5F file

In this example we are creating a file named "sdoc_v1.h5', which can be sent to collaborators

```python
sdb = sdoc.SDOC()
```

```python
sdb.export('sdoc_v1.h5')
```

### Overwrite your database with one that was sent to you

Example of how to overwrite your database with one that a collaborator sent to you (e.g. "sdoc_v1.h5").

This will completelly replace your database with the new one, meaning that any optical constant data that is not present on the new version, will not be saved.

```python
sdb.overwrite('sdoc_v1.h5')
```
